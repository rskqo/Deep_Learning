```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy import ndimage
from skimage.color import rgb2gray
```

```python
img = cv2.imread('/content/drive/MyDrive/ريم سالم  العتيبي/Session 2 /glasses.jpeg')
```

```python
plt.imshow(img)
plt.show()
```

```python
imageRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print('RGB Image:')
plt.imshow(imageRGB)
plt.show()
```

```python
blurG = cv2.GaussianBlur(imageRGB, (9, 9), 50)
plt.imshow(blurG)
plt.show()
```
